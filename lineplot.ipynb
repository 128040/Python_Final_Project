```python
from google.colab import drive
drive.mount('/content/drive')

%cd "/content/drive/My Drive/Colab Notebooks/Python_Project/Python_Data"

import pandas as pd
import matplotlib.pyplot as plt


df_debt = pd.read_csv('Verschuldung_Staat_.csv', delimiter=';')

df_debt_clean = df_debt.copy()

variables_set = ['2014', '2015', '2016', '2017', '2018', '2019', '2020','2021','2022']

df_debt_clean = df_debt_clean[df_debt_clean.columns.tolist()]

df_debt_clean[variables_set] = df_debt_clean[variables_set].replace(",", ".", regex=True).astype("float")

print(df_debt_clean)
```

```python
import pandas as pd

region_map = {
    'Belgium': 'Central Europe',
    'Netherlands': 'Central Europe',
    'Luxembourg': 'Central Europe',
    'Germany': 'Central Europe',
    'Austria': 'Central Europe',
    'Slovenia': 'East Europe',
    'Estonia': 'North Europe',
    'Latvia': 'North Europe',
    'Lithuania': 'North Europe',
    'Czechia': 'East Europe',
    'Slovakia': 'East Europe',
    'Poland': 'East Europe',
    'Hungary': 'East Europe',
    'Bulgaria': 'East Europe',
    'Greece': 'South Europe',
    'Spain': 'South Europe',
    'France': 'South Europe',
    'Croatia': 'East Europe',
    'Italy': 'South Europe',
    'Cyprus': 'South Europe',
    'Ireland': 'North Europe',
    'Denmark': 'North Europe',
    'Finland': 'North Europe',
    'Sweden': 'North Europe',
    'Portugal': 'South Europe',
    'Romania': 'East Europe',
    'Malta': 'South Europe'
}

df_debt_clean['region'] = df_debt_clean['Country'].apply(lambda x: region_map.get(x, 'Other'))

df_debt_clean
```

```python
df_debt_clean_central = df_debt_clean[df_debt_clean['region']=='Central Europe']
df_debt_clean_south = df_debt_clean[df_debt_clean['region']=='South Europe']
df_debt_clean_east = df_debt_clean[df_debt_clean['region']=='East Europe']
df_debt_clean_north = df_debt_clean[df_debt_clean['region']=='North Europe']

print(df_debt_clean_north)
```

```python
import pandas as pd

# Assuming df_debt_clean_central is your DataFrame

# Create the dictionary with nested structure
data = {
    'Country': [
        {'name': country} for country in df_debt_clean_central['Country'].tolist()
    ]
}

# Add year columns as dictionary keys with corresponding values
i = 0  # Initialize index variable outside the loop
for year in df_debt_clean_central.columns[1:]:
    data['Country'][i][year] = df_debt_clean_central[year].tolist()[i]
    i += 1  # Increment index inside the loop

# Print the dictionary
print(data)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a pandas DataFrame from the data
data = {
    'Country': ['Belgium', 'Germany', 'Luxembourg', 'Netherlands', 'Austria'],
    '2014': [107.0, 75.3, 21.9, 67.9, 84.0],
    '2015': [105.2, 71.9, 21.1, 64.7, 84.9],
    '2016': [105.0, 69.0, 19.6, 61.9, 82.8],
    '2017': [102.0, 65.2, 21.8, 57.0, 78.5],
    '2018': [99.9, 61.9, 20.9, 52.4, 74.1],
    '2019': [97.6, 59.6, 22.4, 48.6, 70.6],
    '2020': [111.8, 68.8, 24.6, 54.7, 83.0],
    '2021': [108.0, 69.0, 24.5, 51.7, 82.5],
    '2022': [104.3, 66.1, 24.7, 50.1, 78.4]
}

df_debt = pd.DataFrame(data)

# Melt the DataFrame to prepare for plotting (reshape)
df_melted = df_debt.melt(id_vars='Country', var_name='Year', value_name='Debt')

# Create the line plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
for country in df_debt['Country'].unique():
    subset = df_melted[df_melted['Country'] == country]
    plt.plot(subset['Year'], subset['Debt'], label=country)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Debt')
plt.title('Economic Indicator over Time')

# Rotate x-axis labels for better readability if many years are present
plt.xticks(rotation=45)  # Adjust the rotation angle as needed

# Add legend
plt.legend()

# Show the plot
plt.grid(True)  # Add grid lines for better readability (optional)
plt.tight_layout()  # Adjust spacing between elements
plt.show()

```
